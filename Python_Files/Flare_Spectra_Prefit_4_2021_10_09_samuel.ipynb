```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
# Loads in data file obtained from INSPEX
data = np.loadtxt('4_2021_10_09_samuel.txt',delimiter=',', skiprows=1)
```

```python
# All rows, 2nd column -> x-axis data

energies= data[:,1]
```

```python
# All rows 3rd column -> y-axis data

fluxes = data[:,2]
```

```python
# All rows 4th column -> error data

error = (data[:,3])
#print(error)
```

```python
fig = plt.figure()
ax=plt.axes()
ax.set_xscale("log")
ax.set_yscale("log")
ax.errorbar(energies,fluxes,yerr=error, fmt = 'c')
plt.xlabel('Energy (keV)')
plt.ylabel('Electron Flux ((cm^2 sr s keV)^-1)')
plt.title('Flare Spectra: Flux versus Energy')

# Adds error bars
for count,i in enumerate(list(energies)):
       this_y=list(fluxes)[count]
       this_err=list(error)[count]
       ax.plot([i,i],[this_y-this_err,this_y+this_err],color='k', linestyle='-', linewidth=2)
        
plt.savefig('4_2021_10_09.png', dpi=300)
```
